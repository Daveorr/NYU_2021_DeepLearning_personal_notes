# SGD Optimizer

Stochastic gradient descent (often abbreviated SGD) is an iterative method for optimizing an objective function with suitable smoothness properties (e.g. differentiable or subdifferentiable). 

Considering the "normal" **Full batch gradient descend**:

where:
- $\omega$ are parameters to be updated
- $\eta$ is a learning rate (in more complex cases this can become a positive (semi)definite matrix)
- $S$ is an objective function
- $\frac{\partial{L}(S, \omega)}{\partial{\omega}} = \begin{bmatrix} \frac{\partial{L}(S, \omega)}{\partial{\omega_1}} \\ \vdots \\ \frac{\partial{L}(S, \omega)}{\partial{\omega_n}} \\ \end{bmatrix} $ is the gradient of the objective function with respect of its parameters averaged over all the samples

We update parameters vector $\omega$ as:

$$\omega \leftarrow \omega - \eta\frac{\partial{L}(S, \omega)}{\partial{\omega}}$$

In the **stochastic case** we DO NOT compute the gradient of the whole objective function with respect of its weights like above, but INSTEAD we follow this steps:
- Take one random sample
- Compute the gradient of the objective function for that sample w.r.t. parameters
- Update parameters (take a step (hopefully) downward)
- Repeat until you reach global minimum

The **SGD** formula can be written as:

$$\omega \leftarrow \omega - \eta\frac{\partial{L}(x[p], y[p], \omega)}{\partial{\omega}}$$

The main concepts of SGD are:
- The evaluation that we get from a single sample can be seen as a noisy estimation of the full gradient
- The average of the gradients will be the gradient of the average, meaning that the average trajectory generated by n steps will be close enough to the trajectory that we would obtain with the full batch approach
- SGD is more efficient than full batch in terms of speed of convergence because it can exploit the redundancy in the samples (faster parameters update)
- SGD is often used with $\underline{\textbf{mini-batches}}$, meaning that instead of taking just one random samples we take batches of usually something in between 20 to few thousands samples, we then compute the average of the gradients over the batch and then we update our parameters
- Mini-batch approach is NOT used for an intrinsically algorithmic reason but because it represent a simple way of parallelizing stochastic gradient on parallel hardware such as GPUs

### Example 1: Optimizing a quadratic function with SGD, batch_size = 1

```python
# A downhill movement is made by first calculating how far to move in the input space, calculated as the step size 
# (called alpha or the learning rate) multiplied by the gradient. This is then subtracted from the current point, 
# ensuring we move against the gradient, or down the target function.
# x = x – step_size * f'(x)

class SGDOptimizer(Optim):
    def __init__(self, params, lr):
        print(params)
        defaults = dict(lr=lr)
        super(SGDOptimizer, self).__init__(params, defaults)
        
    def my_step(self, p, state, group):
        lr = group['lr'] # learning rate
        d_p = p.grad.data # derivative of the batch
        print("p: {} - d_p: {}".format(p.data, d_p))
        p.data.add_(-group['lr'], d_p)
```

```python
# please notice, next step would be:
# p - lr * d_p
output(SGDOptimizer([w], lr=0.33333), fname="sgd_lr_opt_example.pdf")
output(SGDOptimizer([w], lr=0.1), fname="sgd_lr_good_example.pdf")
```

We see a lot of variance in the direction of the optimization if we use a large learning rate ($l_r=0.33333$) and an inability to effectively minimise the function if we use a small learning rate ($l_r=0.1$)

# SGD with momentum

SGD with momentum is a method which helps accelerate gradients vectors in the right directions, thus leading to faster converging. It is one of the most popular optimization algorithms and many state-of-the-art models are trained using it.

**Intuition**:

Momentum add's a certain "force" to the direction we were travelling in before. Intuitively, from a physics perspective our optimizer has a certain momentum built in when travelling through our scary loss landscape. The benefits of momentum are as follows:

- Accelerates Optimization. Emphirically this has been observed when we train neural networks. However, no theory confirms this. We can however theoretically confirm this for quadratics.
- Noise Smoothening. Momentum can help skip over local minima and help aboid the draqstic direction changes as we had seen in vanilla SGD.

We define a momentum, which is a moving average of our gradients. We then use it to update the weight of the network as follows:

$$\nu_{(t+1), i} = \beta\nu_{t,i}+\frac{\partial{C}}{\partial{\omega_{t, i}}} \longrightarrow \text{momentum as moving average of the gradient}$$

**SGD with momentum**:

$$\omega_{(t+1),i}=\omega_{t,i}-l_r\nu_{(t+1),i}$$

$\beta$ is a hyperparameter we have to tune. A popular choice of $\beta$ is 0.9 which intuitively says that we care more about the past directions of the gradient than the direction indicated in the current step. 

**Note**: $\nu_{t,1}$ for t=0 is 0

momentum effect in gradient moving average (https://towardsdatascience.com/stochastic-gradient-descent-with-momentum-a84097641a5d)

### Example 2: Optimizing a quadratic function with SGD, batch_size = 1 and momentum

```python
class SGDMomentumOptimizer(Optim):
   
    def __init__(self, params, lr, momentum):
        defaults = dict(lr=lr,momentum=momentum)
        super(SGDMomentumOptimizer, self).__init__(params, defaults)

    def my_step(self, p, state, group):
        lr = group['lr']
        momentum = group['momentum']
        d_p = p.grad.data
        
        if 'v' not in state:
            state['v'] = torch.clone(d_p).detach()
        else:
            state['v'] = d_p + momentum*state['v']
        d_p = state['v']
        print("p: {} - d_p: {} - v: {}".format(p.data, d_p, state['v']))
        p.data.add_(-group['lr']*d_p)
```

## Learning rate and Momentum

An important point to consider when measuring the effect of momentum is to change the learning rate by a similiar amount. This can be achieved by something like this:

$ l_{r(new)} = l_{r(old)}(1 - \beta) $

Unfortunately, tuning these hyper params is a real part of training neural nets effectively :(

```python
output(SGDMomentumOptimizer([w], lr=0.15*(1-0.5), momentum=0.5), nsteps=10, fname="sgd_50mom_example.pdf")
output(SGDMomentumOptimizer([w], lr=0.15*(1-0.25), momentum=0.25), nsteps=10, fname="sgd_25mom_example.pdf")
output(SGDMomentumOptimizer([w], lr=0.15*(1-0.75), momentum=0.75), nsteps=10, fname="sgd_75mom_example.pdf")
```

### Example 3: SGD + momentum with noise

In the below example, we can see a much obvious positive effect of using momentum if we inject noise (increase non convexity) in our loss landscape. Momentum is able to ignore this noise more effectively than vanilla SGD.

```python
output(SGDMomentumOptimizer([w], lr=0.1, momentum=0.0), nsteps=30, noise=1.0, fname="sgd_noise_example.pdf")
output(SGDMomentumOptimizer([w], lr=0.1*(1-0.8), momentum=0.8), nsteps=30, noise=1.0, fname="sgd_mom_noise_example.pdf")
```

# RMSProp Optimizer

**Root Mean Squared Propagation**, or RMSProp, is an extension of gradient descent that uses a decaying average of partial gradients in the adaptation of the step size for each parameter. The effect is to allows the algorithm to forget early gradients and focus on the most recently observed partial gradients seen during the progress of the search.

The intuition is that different layers of a neural network learn at different rates. For example, for a big convolutional net, the output is less sensitive to the pertubations of the weights of the earlier layers and more sensitive to pertubations to the weights of the later layers.

Therefore, in some networks using a larger learning rate for earlier layers and a smaller learning rate for later layers could aid in faster and better convergence.

The central idea of RMSprop is to keep the moving average of the squared gradients for each weight. And then we divide the gradient by square root the mean square. Which is why it’s called RMSprop(root mean square).

$$ v_{t+1, i} = \alpha * v_{t,i} + (1 - \alpha) * (\frac{\partial{C}}{\partial{w_{t,i}}})^2 \longrightarrow \text{moving average of the squared gradients}$$

**RMSProp:**
$$ w_{t+1,i} = w_{t, i} - \frac{l_r}{\sqrt{v_{t+1,i}} + \epsilon} $$

Here, $v$ is simply the moving average of the root mean square of the gradient, $\epsilon$ avoid division by 0.

### Example 4: RMSProp with quadratic function and noise

```python
class RMSPropOptimizer(Optim):
   
    def __init__(self, params, lr, alpha):
        self.eps = 1e-8
        defaults = dict(lr=lr,alpha=alpha)
        super(RMSPropOptimizer, self).__init__(params, defaults)

    def my_step(self, p, state, group):
        lr = group['lr']
        alpha = group['alpha']
        grad = p.grad.data
        
        if 'v' not in state:
            state['v'] = torch.zeros_like(p.data)

        state['v'] = (1-alpha)*torch.pow(grad,2) + alpha*state['v']
        grad.div_(torch.sqrt(state['v']) + self.eps)
        p.data.add_(-group['lr']*grad)
```

```python
output(RMSPropOptimizer([w], lr=0.1, alpha=0.8), nsteps=30, noise=1.0, fname="rmsprop_noise_example.pdf")
```

# Adam Optimizer

The name Adam is derived from **Adaptive Moment Estimation** and is a popular algorithm in the field of deep learning because it achieves good results fast. In the Stanford course on deep learning for computer vision titled “CS231n: Convolutional Neural Networks for Visual Recognition” developed by Andrej Karpathy, et al., the Adam algorithm is again suggested as the default optimization method for deep learning applications.

but what is Adam ? It is simply RMSProp + Momentum.

The root mean square term is calculated exactly like as before:

$$ v_{t+1, i} = \alpha * v_{t,i} + (1 - \alpha) * (\frac{\partial{C}}{\partial{w_{t,i}}})^2 $$ 

The momentum update is given as below, it is roughly equivalent to the momentum in SGD: 

$$ m_{t+1, i} = \beta * m_{t,i} + (1 - \beta) * (\frac{\partial{C}}{\partial{w_{t,i}}}) $$ 

**Adam Optimizer**: 

$$ w_{t+1,i} = w_{t, i} - \frac{lr*m}{\sqrt{v_{t+1,i}} + \epsilon} $$

Just plugging this in will give us `nan` values, to correct for this we do something called **bias scaling**. The idea behind bias scaling is simple. which just keeps the moving averages unbiased during early iterations. The algorithm quickly approaches the above steady state form.

We define the parameters $b_1$ and $b_2$ that control the decay rates of the exponential moving average of the gradient and the of squared gradient.

The equation of Adam with bias scaling is given as follows:
$$ b_1 = (1 - \beta)^t $$$$ b_2 = (1 - \alpha)^t $$

where $t$ is the step number of the gradient update. We use these values to modify our final weight update as follows:

**Adam with bias Scaling**:

$$ w_{t+1,i} = w_{t, i} - lr*\frac{\frac{m}{ b_1}}{\sqrt{\frac{v_{t+1,i}}{b_2}} + \epsilon} $$

where:

- $l_r$. As the learning rate or step size. The proportion that weights are updated (e.g. 0.001). Larger values (e.g. 0.3) results in faster initial learning before the rate is updated. Smaller values (e.g. 1.0E-5) slow learning right down during training
- $b_1$. The exponential decay rate for the first moment estimates (e.g. 0.9).
- $b_2$. The exponential decay rate for the second-moment estimates (e.g. 0.999). This value should be set close to 1.0 on problems with a sparse gradient (e.g. NLP and computer vision problems).
- $\epsilon$. Is a very small number to prevent any division by zero in the implementation (e.g. 10E-8).


```python
import math 

class AdamOptimizer(Optim):
   
    def __init__(self, params, lr, alpha, momentum):
        self.eps = 1e-8

        defaults = dict(lr=lr, alpha=alpha, momentum=momentum)
        super(AdamOptimizer, self).__init__(params, defaults)

    def my_step(self, p, state, group):
        lr = group['lr']
        alpha = group['alpha']
        momentum = group['momentum']
        grad = p.grad.data

        if len(state) == 0:
          state['step'] = 0
          state['m'] = torch.zeros_like(p.data)
          state['v'] = torch.zeros_like(p.data)

        state['step'] += 1
        m, v = state['m'], state['v']

        bias_correction1 = 1 - momentum ** state['step']
        bias_correction2 = 1 - alpha ** state['step']

        m.mul_(momentum).add_(1 - momentum, grad)

        v.mul_(alpha).addcmul_(1 - alpha, grad, grad)

        denom = (v.sqrt() / math.sqrt(bias_correction2)).add_(self.eps)
        step_size = lr / bias_correction1
        p.data.addcdiv_(-step_size, m, denom)
```

As we can see, the updates are much smoother than RMSProp and we get the same benefits of adaptive learning rates !

The situation is not as clear when we train deep neural nets. Remember, for some problems a well tuned SGD + Momentum outperforms many of the state of the art optimization techniques :)

```python
output(AdamOptimizer([w], lr=0.1, alpha=0.7, momentum=0.7), nsteps=30, noise=1.0, fname="adam_noise_example.pdf")
```
